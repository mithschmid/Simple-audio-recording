```python
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
from scipy.io.wavfile import write
import IPython.display as ipd
```

```python
fs = 44100
seconds =  3

print('recording ...')
chunk = sd.rec(frames=int(seconds * fs), samplerate=fs, channels=1)
sd.wait()
print('... finished')
write('output.wav', fs, chunk)
```

```python
plt.figure(figsize=(20, 5))
plt.title('Recorded signal')
t_max = len(chunk) / fs
plt.plot(np.linspace(0, t_max, int(t_max * fs)), chunk)
plt.xlabel('time')
plt.grid()
plt.show()
```

```python
ipd.Audio('output.wav')
```

```python
recordarr = np.array(chunk) # , dtype=float)
```

# Vizualise data

```python
rec_fft = np.fft.fft(recordarr)

rec_freq = np.fft.fftfreq(n=rec_fft.size, d=1/fs)

plt.figure(figsize=(20, 5))
plt.title('FFT of signal')
plt.plot(rec_freq, rec_fft);
plt.xlabel('1/time')
plt.grid()
plt.show()
```
